## Exploratory Data Analysis

### Gender

```python
# Countplot for Gender and it's count:
ax = sns.countplot(x = 'Gender' ,data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index =False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
# Barplot of Gender vs Amount
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

*It can be seen that most of the buyers are female and purchasing power of females are greater then men.*

### Age

```python
# Countplot for Age Group and Gender:
ax = sns.countplot(data = df, x = 'Age Group' ,hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Barplot of Age Group vs Amount:
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount',data = sales_age )
```

 *Most of the buyers lie in age group betweenn 26-35.*

### State

```python
# Barplot of State vs Orders:
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders' )
```

```python
# Barplot of State vs Amount:
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount' )
```

*Most of the Orders and Total Sales are from Uttar Pradesh, Maharashtra and Karnataka respectively.*

### Maritial Status

```python
# Countplot of Marital Status:
ax = sns.countplot(data = df,x = 'Marital_Status')
sns.set(rc ={'figure.figsize':(17,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Barplot of Marital Status vs Amount:
mar_gen = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc ={'figure.figsize':(6,5)})
sns.barplot(data = mar_gen, x ='Marital_Status',y = 'Amount', hue ='Gender')
```

*Most of the buyers are married(women) and they have high purchasing power.*














### Occupation

```python
# Countplot for Occupation:
occ_ax = sns.countplot(data = df, x = 'Occupation')
sns.set(rc = {'figure.figsize':(28,5) } )
for bars in occ_ax.containers:
    occ_ax.bar_label(bars)
```

```python
# Barplot of Occupation vs Amount:
occ_amount = df.groupby(['Occupation'],as_index = False) ['Amount'].sum().sort_values( by = 'Amount' ,ascending = False).head(10)
sns.barplot(data = occ_amount, x = 'Occupation', y = 'Amount'  )
```

 *Most of the buyers are working in IT, Healthcare and Aviation Sector.*

### Product Category

```python
# Countplot for Product Category:
prod_ax = sns.countplot(data = df, x = 'Product_Category')
for bars in prod_ax.containers:
    prod_ax.bar_label(bars)
```

```python
# Barplot of Product Category vs Amount:
prod_amount = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = prod_amount, x = 'Product_Category',y= 'Amount')
```

```python
# Top 10 most sold products:

fig1, ax1 = plt.subplots(figsize=(21,5))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

*Most Products are sold in Food, Clothing and Electronics category.
Clothing has the highest order count but most amount has been spend in Food*

### Conclusion

*Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
