## Data Analysis:

### 1. Demographic Insights:

Gender wise distribution of consumers by Age Group:

```python
# Who Prefers Energy drink More?
sns.countplot(x = df.Gender, data = df, hue = 'Age',saturation=0.55, palette=['#432371','#FAAE7B','#f57542','#9cf542','#42f5a7'])
plt.title('Gender Count')
plt.ylabel('');
```

```python
Age_freq = (df['Consume_frequency'].groupby(df['Age']).value_counts()).unstack()
Age_perc = df.groupby('Age')['Brand_perception'].value_counts().unstack()
```

Brand perception of respondents by Age-Group:

```python
sns.heatmap(Age_perc,cmap='Greens',annot=True,fmt=".1f",linewidth=.5 )
plt.xlabel('Brand Perception')
plt.title('Brand perception of respondents by Age-Group:');
```

Consume frequency of respondents by Age-Group:

```python
sns.heatmap(Age_freq,cmap='Blues',annot=True,fmt=".1f",linewidth=.5 )
plt.xticks(rotation=30) 
plt.xlabel('Consume Frequency');
```

Type of Marketing Channels and their reach by Age-Group:

```python
# Which type of Marketing reaches which age group?
df4 = (df['Marketing_channels'].groupby(df['Age']).value_counts().unstack())
```

```python
plt.figure(figsize = (16,4))
sns.heatmap(df4, cmap='coolwarm',annot=True,fmt=".1f",linewidth=.5)
plt.title(' Marketing Channels by Age-Group');
```

### 3. Consumer Preferences:

Preferred Ingredients of Consumer:

```python
sns.catplot(x="Ingredients_expected", kind="count", data=df, hue = 'Gender')
plt.xlabel('Ingredients expected')
plt.ylabel('')
plt.title('Preferred Ingredients of Consumer');
```

Packaging_preference & Limited_edition_packaging :

```python
pack_pre = (df['Packaging_preference'].groupby(df['Limited_edition_packaging']).value_counts()).unstack()
pack_pre
```

```python
# Count of Packaging Preference:
pack_pre_ = (df['Packaging_preference'].value_counts())
pack_pre_.plot(kind ='bar',color ='#b5b293',edgecolor = "black")
plt.xticks(rotation=45)
plt.title('Packaging preferences of Consumer')
plt.xlabel('Packaging preference');
```

```python
sns.countplot(y= 'Packaging_preference',data = df,hue = 'Limited_edition_packaging',palette='rocket')
plt.ylabel('Packaging Preference')
plt.xlabel('');
```

### 6. Market Leaders:

Brand-wise distribution according to respondent data:

```python
sns.countplot(df.Current_brands, color = "lightgreen",edgecolor = "black")
plt.xlabel('')
plt.style.use('default')
plt.ylabel('Current Brands');
```

```python
plt.figure(figsize =(6,6))
plt.title('Percentage Wise Distriution')
brands = df['Current_brands'].value_counts()
plt.style.use('ggplot') 
plt.pie(brands, labels = brands.index,pctdistance = 0.7,autopct = '%.2f %%');
```

### Scope of Improvements:

Improvements desired from consumers of CodeX who have given Taste Experience less than 3:

```python
df___3 = df.loc[(df['Taste_experience'] > 3) & (df['Current_brands'] == 'CodeX')]
plt.figure(figsize= (12,5))
sns.countplot(x = df___3.Improvements_desired,color ='yellow',saturation = 0.70,edgecolor = "black")
plt.ylabel('')
plt.xlabel('Improvements desired');
```

Consumers who have prevented CodeX Brand and their reason for doing so

```python
df3not = df.loc[(df['Brand_perception'] != 'Positive') & (df['Current_brands'] == 'CodeX')]
plt.figure(figsize= (12,5))
sns.countplot(x = 'Reasons_preventing_trying', data = df3not,color ='#b4eb34',edgecolor = "black")
plt.ylabel('');
```

Reasons for choosing brands according to respondents:

```python
mar_lea = (df['Reasons_for_choosing_brands'].groupby(df['Current_brands']).value_counts()).unstack()
sns.heatmap(mar_lea,cmap='flare',annot=True,fmt=".1f",linewidth=.5 )
plt.xticks(rotation=45)
plt.xlabel('Reasons for choosing brands')
plt.ylabel('Current brands');
```

Reasons for choosing CodeX Brand:

```python
Codex = df[(df['Current_brands']== 'CodeX')]
Codex_ = (Codex['Current_brands'].groupby(Codex['Reasons_for_choosing_brands']).value_counts()).unstack()
Codex_.plot(kind ='bar',color ='#34ebd3',edgecolor = "black", legend =None )
plt.xticks(rotation=45)
plt.xlabel('Reasons for choosing brands');
```
